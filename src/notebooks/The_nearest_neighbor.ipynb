```python
from data_matrix import cities, distance_matrix, cities2, distance_matrix2
```

```python
def nearest_neighbor(dist_matrix, start=0):
    n = len(dist_matrix)
    visited = [False] * n
    path = [start]
    visited[start] = True
    total_distance = 0
    
    for _ in range(n - 1):
        last = path[-1]
        nearest = None
        nearest_dist = float("inf")
        
        for j in range(n):
            if not visited[j] and dist_matrix[last][j] < nearest_dist:
                nearest = j
                nearest_dist = dist_matrix[last][j]
        
        path.append(nearest)
        visited[nearest] = True
        total_distance += nearest_dist
    
    total_distance += dist_matrix[path[-1]][start]
    path.append(start)
    
    return path, total_distance
```

```python
print("Smaller matrix results")
for idx in range(len(cities)):
    path, dist = nearest_neighbor(distance_matrix, start=idx)

    print("starting city:", cities[idx])
    print("The best route:", " -> ".join([cities[i] for i in path]))
    print("Length:", dist)
    print("-------------------")


print("\nLarger matrix results")
for idx in range(len(cities2)):
    path, dist =  nearest_neighbor(distance_matrix2, start=idx)

    print("starting city:", cities2[idx])
    print("The best route:", " -> ".join([cities2[i] for i in path]))
    print("Length:", dist)
    print("-------------------")
```
