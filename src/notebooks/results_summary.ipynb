```python
from ipynb.fs.full.genetic_algorithm import genetic_algorithm
from ipynb.fs.full.bound_and_branch import branch_and_bound_tsp
from ipynb.fs.full.brute_force_algoritam import brute_force_tsp
from ipynb.fs.full.hungarian_metod import Hungarian1, preprocess_matrix, merge_cycles_into_tour, Hungarian2, merge_cycles_into_tour2
from ipynb.fs.full.The_nearest_neighbor import nearest_neighbor
from ipynb.fs.full.genetic_algorithm import genetic_algorithm
from data_matrix import cities, distance_matrix, cities2, distance_matrix2, distance_matrix3, cities3
import time
import pandas as pd
import numpy as np
import math
INF = math.inf
```

```python
# Brute force
start = time.perf_counter()
result = brute_force_tsp(distance_matrix, cities)
end = time.perf_counter()
brute_force_path, brute_force_cost = result["best_route_cities"], result["best_length"]
brute_force_time = end - start

# Bound and Branch
cost_matrix = [ [INF, 12, 10, 19, 8, 14],
                [12, INF, 3, 7, 2, 11],
                [10, 3, INF, 6, 20, 4],
                [19, 7, 6, INF, 13, 9], 
                [8, 2, 20, 13, INF, 16], 
                [14, 11, 4, 9, 16, INF] ]

start = time.perf_counter()
bb_path, bb_cost = branch_and_bound_tsp(cost_matrix)
end = time.perf_counter()
bb_path = [cities[i] for i in bb_path]

bb_indexes = [cities.index(city) for city in bb_path]
bb_cost2 = 0
for i in range(len(bb_indexes) - 1):
    bb_cost2 += distance_matrix[bb_indexes[i]][bb_indexes[i + 1]]
bb_cost2 += distance_matrix[bb_indexes[-1]][bb_indexes[0]]
bb_time = end - start

# Hungarian
start = time.perf_counter()
solver1 = Hungarian1(preprocess_matrix(distance_matrix))
assignment1 = solver1.solve()
end = time.perf_counter()
merged_tour1 = merge_cycles_into_tour(assignment1, distance_matrix)
hungarian_cost = sum(distance_matrix[i][j] for i, j in assignment1)
hungarian_cost += distance_matrix[merged_tour1[-1]][merged_tour1[0]]
hungarian_path = [cities[x] for x in merged_tour1] + [cities[merged_tour1[0]]]
hungarian_time = end - start

# Hungarian 2
start = time.perf_counter()
solver2 = Hungarian2(distance_matrix)
assignment2 = solver2.solve()
end = time.perf_counter()
n = len(cities)
merged_tour2 = merge_cycles_into_tour2(assignment2, distance_matrix, n)
hungarian_path2 = [cities[x] for x in merged_tour2] + [cities[merged_tour2[0]]]
hungarian_cost2 = 0
for i in range(len(merged_tour2) - 1):
    hungarian_cost2 += distance_matrix[merged_tour2[i]][merged_tour2[i + 1]]
hungarian_cost2 += distance_matrix[merged_tour2[-1]][merged_tour2[0]]
hungarian2_time = end - start

# Nearest Neighbor
start = time.perf_counter()
nn_path, nn_cost = nearest_neighbor(distance_matrix)
end = time.perf_counter()
nn_path = [cities[i] for i in nn_path]
nn_time = end - start

# Genetic Algorithm
start = time.perf_counter()
ga_path, ga_cost, best_dist_per_gen, avg_dist_per_gen= genetic_algorithm(distance_matrix, cities)
end = time.perf_counter()
ga_time = end - start

# --- Upisivanje u tabelu ---
results = pd.DataFrame([
    {"Algorithm": "Brute Force", "Path": brute_force_path, "Cost": brute_force_cost, "Time (s)": brute_force_time},
    {"Algorithm": "Bound and Branch", "Path": bb_path, "Cost": bb_cost2, "Time (s)": bb_time},
    {"Algorithm": "Hungarian Method", "Path": hungarian_path, "Cost": hungarian_cost, "Time (s)": hungarian_time},
    {"Algorithm": "Hungarian Method2", "Path": hungarian_path2, "Cost": hungarian_cost2, "Time (s)": hungarian2_time},
    {"Algorithm": "Nearest Neighbor", "Path": nn_path, "Cost": nn_cost, "Time (s)": nn_time},
    {"Algorithm": "Genetic Algorithm", "Path": ga_path, "Cost": ga_cost, "Time (s)": ga_time},
])

df = pd.DataFrame(distance_matrix, index=cities, columns=cities)
print(df)
print('\n')
print(results)
```
